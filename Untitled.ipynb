# Visualization to Identify Trends/Patterns

```python
# The first two were are simple enough to do, just doing a easy plot to compare the pricing and the other values
plt.scatter(df['type'], df['price'], color='red')
plt.xlabel("Type of Housing")
plt.ylabel("Rent Price")
plt.xticks(rotation=90)
plt.show()


plt.scatter(df['furnishing'], df['price'], color='blue')
plt.xlabel("Furnished or Unfurnished Housing")
plt.ylabel("Rent Price")
plt.show()

# Since I had no idea where the error were coming from and I had to many data, I simply transformed the 'baths' columns
# into a numeric one and denied the incoming error feedbacks
df['baths'] = pd.to_numeric(df['baths'], errors='coerce')
plt.scatter(df['baths'], df['price'], color='black')
plt.xlabel("Number of Bathroom or Water Room")
plt.ylabel("Rent Price")
plt.show()

# For this one , the str.extract wouldn't work. So I had to convert the values into full str to be able to extract the numerical
# values of it and than converts the values to float.
df['beds'] = df['beds'].astype(str)
df['beds_numeric'] = df['beds'].str.extract('(\d+)').astype(float)
df = df.dropna(subset=['beds_numeric', 'price'])
plt.figure(figsize=(10, 6))
plt.scatter(df['beds_numeric'], df['price'], color='green')
plt.xlabel("Number of Bedrooms / Beds per Housing")
plt.ylabel("Rent Price")
plt.show()
```

# Data Cleaning and Preparation

```python
y = df['price']
X = df.drop("price", axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X= X.dropna(axis=1)
```

#  Feature Engineering

I'll do one code of each model with the scaling and CV while the other one without it to see and understand the difference

#### Lasso / L1 Regularization

```python
#Load packages 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score

#Training my model with 85% of the data set and using the remaining 15% for the test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

#Proceeding to use lasso model with a manually picked alpha ( I tried 1, 10 , 100 ) this one was the best one out of the 4
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

mse_lasso, mae_lasso, r2_lasso
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)



lassocv_model = LassoCV(eps=0.01,n_alphas=200,cv=10)
lassocv_model.fit(scaled_X_train, y_train)
lassocv_model.alpha_
```

We can compare the 0.1 alpha to the 3,7544 and make a distinction between the two method to proceed with the Alpha

```python
y_pred_lasso2 = lassocv_model.predict(scaled_X_test)
mse_lasso2 = mean_squared_error(y_test, y_pred_lasso2)
mae_lasso2 = mean_absolute_error(y_test, y_pred_lasso2)
r2_lasso2 = r2_score(y_test, y_pred_lasso)

mse_lasso2, mae_lasso2, r2_lasso2
```

#### Ridge / L2 Regularization

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.5)
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

mse_ridge, mae_ridge, r2_ridge

# I pretty much did the same thing I did with lasso but now I'll try to use RidgeCV below
```

I pretty much did the same thing I did with lasso but now I'll try to use RidgeCV below

```python
from sklearn.linear_model import RidgeCV

ridgecv_model = RidgeCV((0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')

ridgecv_model.fit(scaled_X_train, y_train)

ridgecv_model.alpha_
```

We can see that alpha 10 is better than the 0.5 I gave it earlier so we will redo the model with an alpha of 10 and see the difference

```python


y_pred_ridge2 = ridgecv_model.predict(scaled_X_test)

MAE = mean_absolute_error(y_test, y_pred_ridge2)
MSE = mean_squared_error(y_test, y_pred_ridge2)
r2_ridge2 = r2_score(y_test, y_pred_ridge2)


MAE, MSE, r2_ridge2
```

The 0.004 difference isn't worth talking about but it's still information

#### Tree-Based / Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor


tree = DecisionTreeRegressor(random_state=42, max_depth=500)
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)
mse_tree = mean_squared_error(y_test, y_pred_tree)
mae_tree = mean_absolute_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)

mse_tree, mae_tree, r2_tree
```

# Model Comparison

#### Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
```

```python
models = {
    'Lasso': Lasso(alpha=0.1, max_iter = 1000),
    'Ridge': Ridge(alpha=1.0),
    'DecisionTree': DecisionTreeRegressor()
}

results = {}
for model_name, model in models.items():
    scores = cross_validate(model, X, y, cv=5, scoring='r2', return_train_score=False)
    results[model_name] = {
        'Mean Test Score': np.mean(scores['test_score']),
        'Std Test Score': np.std(scores['test_score'])
    }
    
results_df = pd.DataFrame(results).T
print(results_df)
```
